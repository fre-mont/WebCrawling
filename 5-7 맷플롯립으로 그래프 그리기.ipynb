```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```python
x =[1,4,9,16,25,36,49,64]
plt.plot(x)
plt.show(x)
```

```python
# 색깔 바꾸기
plt.plot(x, color='red')
```

```python
plt.plot(x, 'or')
```

```python
y = [i for i in range(1,9)]
plt.plot(x, y)
# 축 이름 지정하기
plt.xlabel('x')
plt.ylabel('y')
plt.title("matplotlib sample", color='white')
plt.show()
```
